```python
from pathlib import Path

data_dir = Path("..") / "data"

list(data_dir.glob("*"))
```

```python
import pandas as pd
```

```python
train_values = pd.read_csv(data_dir / "train_values.csv").set_index("building_id")

train_values
```

```python
train_values.dtypes
```

```python
train_labels = pd.read_csv(data_dir / "train_labels.csv").set_index("building_id")
train_labels
```

```python
test_values = pd.read_csv(data_dir / "test_values.csv").set_index("building_id")
test_values
```

```python
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder

import xgboost
```

```python
features = ColumnTransformer(
    [('scale', StandardScaler(), make_column_selector(dtype_include=np.number)),
     ('categories', OneHotEncoder(drop="first"), make_column_selector(dtype_exclude=np.number))]
)

features.fit(train_values)
```

```python
pl = Pipeline([
    ('features', features),
    ('clf', RandomForestClassifier())
])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = [{'clf__max_depth': [10, 25],
           'clf__n_estimators': [50]
}, {
    'clf': (xgboost.XGBClassifier(), ),
    'clf__max_depth': [10, 25],
    'clf__learning_rate': [1e-3, 1e-2],
    'clf__n_estimators': [25, 50],
    'clf__use_label_encoder': [False]
}]

est = GridSearchCV(pl, params, verbose=3, scoring="f1_micro")
est.fit(train_values, train_labels.values.ravel() - 1)
```

```python
pd.DataFrame(est.cv_results_) 
```

```python
params = [
    {
        'clf__max_depth': [38, 30],
        'clf__n_estimators': [50]
}]

est = GridSearchCV(pl, params, verbose=3, scoring="f1_micro", n_jobs=4)
est.fit(train_values, train_labels.values.ravel() - 1)
```

```python
pd.DataFrame(est.cv_results_)
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = [{
    'clf': (xgboost.XGBClassifier(eval_metric=["merror", "map", "auc"], objective="multi:softmax"), ),
    'clf__max_depth': [5, 30, 25, 35],
    'clf__learning_rate': [3e-1, 2e-1, 1e-1, 1e-2],
    'clf__n_estimators': [300, 100, 40],
    'clf__reg_alpha': [0, 0.5, 1],
    "clf__gamma" : [0.01, 0.1, 0.3, 0.5, 1, 1.5, 2],
    "clf__reg_lambda": [1, 1.5, 2, 3, 4.5],
    "clf__min_child_weight": [1, 10, 30],
    "clf__subsample": [0.2, 0.4, 0.5, 0.6, 0.7],
    'clf__use_label_encoder': [False]
}]

est = RandomizedSearchCV(pl, params, verbose=3, scoring="f1_micro", n_iter=20)
est.fit(train_values, train_labels.values.ravel() - 1)
```

```python
prediction = pd.Series(est.predict(test_values), index=test_values.index).rename('damage_grade') + 1
prediction
```

```python
prediction.to_csv("../data/submissions/submission_xgb_randomizedsearch.csv")
```
